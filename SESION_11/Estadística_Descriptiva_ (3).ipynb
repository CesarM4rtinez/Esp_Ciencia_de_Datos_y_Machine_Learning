#### 5. Media Geomtrica


```python
import numpy as np  # K = platicurtica

def media_geometrica(iterable):
  ''' Calcula la media geomtrica'''
  a = np.log(iterable)
  return "Media Geomtrica", np.exp(a.sum()/len(a)).round(2)

```

```python
df[var_num].apply(media_geometrica, axis=0).T
```

#### 7. Media Recortada

Al 20%; 10% en cada lado



```python
from scipy import stats as nstats

def media_recortada_20(x):
  mr = nstats.trim_mean(x, proportiontocut=0.10).round(2) # 20%
  return "Media recortada al 20%", mr
```

```python

df[var_num].apply(media_recortada_20, axis=0).T
```

#### 7. Trimedia

```python
def trimedia(x: pd.Series):
    #Obtenemos los "Quartiles"
    Q1 = x.quantile([.25]).values[0]
    Q2 = x.quantile([.50]).values[0]
    Q3 = x.quantile([.75]).values[0]
    t = (Q1 + 2*Q2 + Q3) / 4
    return t
```

```python

df[var_num].apply(trimedia, axis=0).round(2)
```

###   B.     MEDIDAS DE DISPERSION

#### 1. Rango

```python
def rango(x):
  min = np.min(x)
  max = np.max(x)
  r = max - min
  return min, max, r
```

```python
d = df[var_num].apply(rango, axis=0).T
d.columns = ["Mnimo","Mximo","Rango"]
d
```

#### 2. Cuartiles

```python
def cuartiles(x: pd.Series):
    #Obtenemos los "Quartiles"
    Q1 = x.quantile([.25]).values[0]
    Q2 = x.quantile([.50]).values[0]
    Q3 = x.quantile([.75]).values[0]
    min = np.min(x)
    max = np.max(x)
    return min, Q1, Q2, Q3, max
          # 0    1   2   3   4
```

```python
#Obtenemos los Quartiles
d = df[var_num].apply(cuartiles, axis=0).T
d.columns = ["Mnimo","P25% - Q1","P50% - Q2", "P75% - Q3", "Mximo"]
d
```

#### 3. Rango intercuartil

```python

def rango_intercuartil(x: pd.Series): # RIC
    #Obtenemos los "Quartiles"
    Q1 = x.quantile([.25]).values[0]
    Q3 = x.quantile([.75]).values[0]

    return Q1, Q3 - Q1, Q3
```

```python
#Obtenemos los Quartiles

d= df[var_num].apply(rango_intercuartil, axis=0).T
d.columns = ["Q1","RIC","Q3"]
d
```

#### 8. Curtosis

```python
from scipy.stats import kurtosis

def kurtosis_data(x):
  curtosis =  kurtosis(x, fisher=True)

  if curtosis >0:
    return ('Es Leptocrtica-->', round(curtosis,2 ))
  elif curtosis ==0: # 0.03
    return  ('Es Mesocrtica-->', round(curtosis,2 )) # curva semejante a la normal
  else:
    return  ('Es Platicrtica-->', round(curtosis,2 ))
```

```python
df[var_num].apply(kurtosis_data, axis=0).T
```

## Correlacin: r de pearson

```python
correlacion = df[var_num].corr(method='pearson').round(3)  # paramtrico

correlacion
```

```python
mask = np.triu(np.ones_like(correlacion, dtype=bool )) #la mscara para el tringulo superior del grfico
f, ax = plt.subplots(figsize = (10,10)) # la base para el grfico
g = sns.heatmap(correlacion, mask = mask, cmap='Blues', vmax=1, center = 0,
            square = True, linewidths=.1, annot=True)

g.set_title("Grfico N XX: Grfico de la correlacin de las variables numricas")
```

```python
g = sns.JointGrid(x="ingreso", y="score", data=df)

g.plot_joint(sns.scatterplot, s=100, alpha=.5)  # Efectos de las bubujas s=100, alpha=.5
g.plot_marginals(sns.histplot, kde=True)
```

```python
# con lnea de tendencia
g=sns.jointplot(x=df['ingreso'], y=df['score'])

sns.regplot(data=df, x= 'ingreso', y='score', ax=g.ax_joint, scatter=False, color= 'red')
plt.show();
```

```python
# con lnea de tendencia por grupos

sns.set(style="ticks")

sns.lmplot(x="ingreso", y="score", col='nivel', data=df,
               col_wrap=2, ci=None, palette="muted", height=4,
               scatter_kws={"s": 50, "alpha": 1});
```
