```python
# Aadiendo variables y clculos con Apply
import numpy as np
import pandas as pd

np.random.seed(123452)

df2 = pd.DataFrame(             #  F, C
    data=np.random.randint(0, 21, (10, 3)) ,    # valores entre 0 y 10, en 10 filas y 5 columnas
    columns=["a", "b", "c"]
    )

df2 = (     # aadiendo variables con lambda
    df2
    .assign(c3 = lambda x: x.a**2 + 10*x.b - 20)
    .assign(c4 = lambda x: x.b*x.c)
    .assign(c5 = lambda x: np.sqrt(x.c).round(3))
    .assign(c6 = lambda x: round((x.a + x.b + x.c)/3, 3) )

)

df2
```

### Tipo: Outer

```python
# generamos data para nuestros ejemplos
df1 = df2.sample(4)
df1 = df1.reset_index(drop=True)   # 3 * 7
df1['ID'] = ["id2","id3", "id5", "id6"]
df1
```

```python
import numpy as np

df2 = pd.DataFrame(
    np.random.randint(1,11, 25).reshape([5, 5]),
    columns = ["b", "c", "d", "e", "h"])

df2['ID'] = ["id1","id3", "id5", "id7", "id9"]
df2  # 3 * 4
```

Colocamos **sort = False** para evitar que se muestre un aviso al respecto de cierto cambio de funcionalidad que se produzaca en futuras versiones de Pandas.

```python
# Unimos los dos dataframes segn la etiqueta de las columnas
pd.concat([df1, df2], sort= False)
```

```python
# Unimos los dos DataFrame's a lo largo de la columna
# Segn el Index

pd.concat([df1, df2], axis = 1, sort = False)
```

```python
# Unimos los dos DataFrame's a lo largo de la columna
# Segn 1 o varias columnas

df1.merge(df2, on=["ID"], how='outer')
```

### Tipo: Inner

```python
# Interseccin (y) a nivel de columnas
pd.concat([df1, df2], join = "inner")  # Slo las columnas que son comunes
```

```python
# Interseccin (y) a nivel de filas
pd.concat([df1, df2], axis = 1, join = "inner") #, ignore_index=True) # slo las filas que son comunes
```

```python
# Unimos los dos DataFrame's a lo largo de la columna (i)
# Segn 1 o varias columnas

df1.merge(df2, on=["ID"], how='inner')
```

### Tipo: Left

```python
# Unimos los dos DataFrame's: Base DataFrame Izquierdo
# Segn 1 o varias columnas

df1.merge(df2, on=["ID"], how='left')
```

### Tipo: Right

```python
# Unimos los dos DataFrame's: Base DataFrame Derecho
# Segn 1 o varias columnas

df1.merge(df2, on=["ID"], how='right')
```

# Datos Duplicados

El mtodo duplicated de la clase DataFrame permite identificar filas repetidas; cuyo resultado es de tipo **Bool**, donde el valor **True** indica que ya existe, caso contrario ser **False** .

datos duplicados.jpg

```python
import pandas as pd

df7 = pd.DataFrame([['Per', 10200],  # generando data para nuestros ejemplos
       ['Brasil', 11506],
       ['Brasil', 11506],
       ['Chile', 5500],
       ['Per', 10200],
       ['Chile', 23],
       ['Ecuador', 10000],
       ['Per', 2560]],
       columns = ['Pais', 'Cifra'])
df7
```

```python
df7.duplicated() #Se est revisando a nivel de todas las columnas
```

```python
# Extraer en un nuevo DataFrame todos los duplicados
df7[df7.duplicated()] # True
```

```python
# Eliminar datos duplicados
df7.drop_duplicates() # nos deja el DataFrame sin los datos duplicados
```

```python
df7  # si se desea mantener el cambio, almacenar en un contenedor
```

```python
# Eliminar duplicados por ciertas columnas en especfico

df7.drop_duplicates(subset = ['Pais'])  # mantuvo el 1ro y elimin el resto de duplicados
```

```python
# Eliminar todas las filas repetidas, pero no la ltima "keep"

df7.drop_duplicates(subset = ['Pais'], keep = 'last')
```
