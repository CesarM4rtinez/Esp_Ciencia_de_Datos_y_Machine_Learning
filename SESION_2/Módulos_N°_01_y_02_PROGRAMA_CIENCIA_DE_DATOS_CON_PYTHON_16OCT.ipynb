# Diccionarios

Es una coleccin de elementos, donde cada uno tiene una llave "key" y un valor "value".

Los diccionarios se pueden crear con llave {} separando con una coma cada par key:value.

```python
dic1 = {} # un diccionario vaco
dic1
```

```python
dicVehiculo = {
  "marca": ["Ford", "Jeep", "Nissan"],
  "modelo": "Mustang",
  "ao": [1964, 1965, 1985]
}

dicVehiculo
```

```python
dicVehiculo['ao'] # obteniendo el valor de un "Key"
```

```python
dicVehiculo['marca']
```

```python
# Si el key al que accedemos no existe, se aade automticamente.
#             Key                Value
dicVehiculo['nro asientos'] = [5, 4, 4, 3, 6, 12]
dicVehiculo
```

```python
dicVehiculo['nro asientos'] = (10, 23, [20])  # actualizando / modificando  un valor asociado a un key
dicVehiculo
```

## Recorriendo los diccionarios

Podemos recorrer de manera similar que a las Tuplas y Listas.

```python
l = list() # lista vaca

for k in dicVehiculo: #imprimir los "Key"
    print(k)
    l.append(k)

print(l)  # lista con los Key del diccionario
```

```python
l = list()

for v in dicVehiculo:  # key
    print(dicVehiculo[v]) #imprimir los valores del Diccionario

    l.append(dicVehiculo[v])
l
```

```python
# Imprime los key y value del diccionario
# diccionario key:valor ---> elemento de un diccionario

for k, v in dicVehiculo.items(): # Key  y Value
    print("Key: " + str(k) + " // Valor: " + str(v))
```
