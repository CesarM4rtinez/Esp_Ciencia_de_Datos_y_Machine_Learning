## Introduccin

La agrupacin en clsteres pertenece a las tcnicas de aprendizaje automtico no supervisadas.

La tarea principal de la agrupacin en clsteres es descubrir grupos "naturales" en un conjunto de datos sin etiquetar. Esta es una tarea importante en el anlisis de datos, ya que se utiliza en muchas aplicaciones cientficas, de ingeniera y comerciales. La aplicacin ms conocida de la agrupacin en clsteres es la segmentacin de clientes (para un marketing eficiente), la segmentacin de imgenes y la agrupacin de documentos.
Hay muchos algoritmos de agrupamiento que se pueden dividir en dos tipos principales: jerrquicos y particionales.

* Los algoritmos <u> jerrquicos </u> dividen de forma recursiva un conjunto de datos en un subconjunto ms pequeo hasta que un subconjunto contiene solo un elemento. Esto se puede representar con un dendrograma que parece un rbol. Puede construirse desde las hojas hasta el suelo (enfoque aglomerativo) o desde la raz hasta las hojas (enfoque divisivo). En la agrupacin jerrquica, no es necesario que especifique la cantidad de agrupaciones, pero debe definir una condicin de terminacin para el proceso de divisin / fusin. </li>

* <u>Partitional</u> Los algoritmos dividen un conjunto de datos en varios subconjuntos (grupos) segn un criterio determinado. Para algunos algoritmos, el nmero de clsteres debe definirse a priori (por ejemplo, K-Means) y para otros no (DBSCAN). Definir el nmero de clsteres antes de ejecutar un algoritmo a menudo requiere un conocimiento de dominio especfico que a menudo es un desafo (o incluso imposible) en muchas aplicaciones. Esto condujo al desarrollo de muchas heursticas y enfoques simplificados que ayudaron a los analistas sin conocimiento del dominio a elegir el nmero apropiado de clsteres. </li>

Existe una gran cantidad de algoritmos de agrupamiento y, actualmente, no hay uno que domine a los dems. La eleccin del mejor depende de la base de datos en s, del dominio de la aplicacin y de los requisitos y expectativas del cliente.
Este cuaderno se centra en tres algoritmos de particin: K-Means, DBSCAN y Affinity Propagation. Todos estn implementados en una conocida biblioteca de Python: Scikit-Learn.

## 1 - K-Means

**K-means** es la implementacin ms simple del principio de mxima separacin y mxima cohesin interna. Supongamos que tenemos un conjunto de datos $X\in ^{MN}$ (es decir, $M N$ muestras dimensionales) que queremos dividir en $K$ grupos y un conjunto de $K$ centroides correspondientes a las medias de las muestras asignadas a cada grupo $K_j$:

$$M^{(0)}=\{\bar \mu_{0}^{(0)}, \bar \mu_{1}^{(0)}, \cdots, \bar \mu_{K}^{(0)}   \} $$
donde
$$ \bar \mu_{i}^{(t)} \in \mathbb{R}^N $$

El conjunto $M$ y los centroides tienen un ndice adicional (como un superndice) que indica el paso iterativo. A partir de una suposicin inicial $M^{(0)}$, K-means intenta minimizar una funcin objetivo llamada inercia (es decir, la distancia total promedio dentro del grupo entre las muestras asignadas a un grupo $K_j$ y su centroide $\mu_j$):
$$S(t)= \sum_{k=1}^{K} \sum_{\bar x_i\in K_j} ||\bar x_i -\bar \mu_{k}^{t} ||^2 $$

Es fcil entender que $S(t)$ no se puede considerar como una medida absoluta porque su valor est muy influenciado por la varianza de las muestras.
Sin embargo, $S(t+1) < S(t)$ implica que los centroides se estn acercando a una posicin ptima donde los puntos asignados a un grupo tienen la menor distancia posible al centroide correspondiente. Por lo tanto, el procedimiento iterativo (tambin conocido como algoritmo de Lloyd) comienza inicializando $M^{(0)}$ con valores aleatorios. El siguiente paso es la asignacin de cada muestra $x_i\in X$ al conglomerado cuyo centroide tiene la menor distancia de $x_i$:

$$c(\bar x_i;M^{(t)})=argmin_{j}^{(t)} d(\bar x_i,\bar \mu_{j}^{(t)}) $$

Una vez que se han completado todas las asignaciones, los nuevos centroides se vuelven a calcular como medias aritmticas:

$$\bar \mu_{j}^{(t)}=\frac{1}{N_{k_j}} \sum_{\bar x_i\in K_j}= \langle\bar x_j \rangle K_j $$

El procedimiento se repite hasta que los centroides dejen de cambiar (esto implica tambin una secuencia $S(0) > S(1) > ... > S(t_{end})$). El lector debera haber entendido de inmediato que el tiempo de clculo est muy influenciado por la suposicin inicial. Si $M^{(0)}$ est muy cerca de $M^{(t_{end})}$, algunas iteraciones pueden encontrar la configuracin ptima. Por el contrario, cuando $M^{(0)}$ es puramente aleatorio, la probabilidad de una eleccin inicial ineficiente es cercana a 1 (es decir, cada eleccin aleatoria uniforme inicial es casi equivalente en trminos de complejidad computacional).

# 2 - Mini Batch K-Means

El `MiniBatchKMeans()` es una variante del algoritmo `KMeans` que utiliza mini-batches para reducir el tiempo de clculo, mientras intenta optimizar la misma funcin objetivo. Los mini-batches son subconjuntos de los datos de entrada, muestreados aleatoriamente en cada iteracin de entrenamiento. Estos mini-batches reducen drsticamente la cantidad de cmputo requerida para converger a una solucin local. A diferencia de otros algoritmos que reducen el tiempo de convergencia de k-means, k-means de mini-batches produce resultados que, por lo general, son ligeramente peores que el algoritmo estndar.

El algoritmo itera entre dos pasos principales, similar a vanilla k-means. En el primer paso, las muestras se extraen aleatoriamente del conjunto de datos para formar un mini-batches. Luego se asignan al centroide ms cercano. En el segundo paso, se actualizan los centroides. A diferencia de k-means, esto se hace por muestra. Para cada muestra en el mini-batches, el centroide asignado se actualiza tomando el promedio de transmisin de la muestra y todas las muestras anteriores asignadas a ese centroide. Esto tiene el efecto de disminuir la tasa de cambio de un centroide con el tiempo. Estos pasos se realizan hasta que se alcanza la convergencia o un nmero predeterminado de iteraciones.

MiniBatchKMeans converge ms rpido que KMeans, pero se reduce la calidad de los resultados.

## Coeficiente y visualizacin silhouette

El coeficiente de silueta se utiliza cuando se desconoce la verdad sobre el terreno del conjunto de datos y calcula la densidad de los clsteres calculados por el modelo. El puntaje se calcula promediando el coeficiente de silueta para cada muestra, calculado como la diferencia entre la distancia promedio dentro del grupo y la distancia media del grupo ms cercano para cada muestra, normalizada por el valor mximo. Esto produce una puntuacin entre 1 y -1, donde 1 son clsteres muy densos y -1 es un agrupamiento completamente incorrecto.

El visualizador de silueta muestra el coeficiente de silueta para cada muestra por grupo, visualizando qu grupos son densos y cules no. Esto es especialmente til para determinar el desequilibrio de grupos o para seleccionar un valor para K comparando varios visualizadores.

El Coeficiente de Silueta se define para cada muestra y se compone de dos puntuaciones:

* a: La distancia media entre una muestra y todos los dems puntos de la misma clase.

* b: La distancia media entre una muestra y todos los dems puntos en el siguiente grupo ms cercano.

El Coeficiente de silueta s para una sola muestra se da como:

$$s = \frac{b - a}{max(a, b)}$$

### Ventajas y Desventajas

**Ventajas**
* La puntuacin est limitada entre -1 para un agrupamiento incorrecto y +1 para un agrupamiento muy denso. Las puntuaciones alrededor de cero indican grupos superpuestos.
* La puntuacin es ms alta cuando los conglomerados son densos y bien separados, lo que se relaciona con un concepto estndar de conglomerado.

**inconvenientes**
* El coeficiente de silueta es generalmente ms alto para los clsteres convexos que para otros conceptos de clsteres, como los clsteres basados en la densidad como los obtenidos a travs de DBSCAN.
