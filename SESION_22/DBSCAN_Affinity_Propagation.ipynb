## 1. Introduction  <a href='#Top' style="text-decoration: none;">^</a><br>

La agrupacin en clsteres pertenece a las tcnicas de aprendizaje automtico no supervisadas.

La tarea principal de la agrupacin en clsteres es descubrir grupos "naturales" en un conjunto de datos sin etiquetar. Esta es una tarea importante en el anlisis de datos, ya que se utiliza en muchas aplicaciones cientficas, de ingeniera y comerciales. La aplicacin ms conocida de la agrupacin en clsteres es la segmentacin de clientes (para un marketing eficiente), la segmentacin de imgenes y la agrupacin de documentos.
Hay muchos algoritmos de agrupamiento que se pueden dividir en dos tipos principales: jerrquicos y particionales.

* Los algoritmos <u> **jerrquicos** </u> dividen de forma recursiva un conjunto de datos en un subconjunto ms pequeo hasta que un subconjunto contiene solo un elemento. Esto se puede representar con un dendrograma que parece un rbol. Puede construirse desde las hojas hasta el suelo (enfoque aglomerativo) o desde la raz hasta las hojas (enfoque divisivo). En la agrupacin jerrquica, no es necesario que especifique la cantidad de agrupaciones, pero debe definir una condicin de terminacin para el proceso de divisin / fusin. </li>

* <u>**Partitional**</u> Los algoritmos dividen un conjunto de datos en varios subconjuntos (grupos) segn un criterio determinado. Para algunos algoritmos, el nmero de clsteres debe definirse a priori (por ejemplo, K-Means) y para otros no (DBSCAN). Definir el nmero de clsteres antes de ejecutar un algoritmo a menudo requiere un conocimiento de dominio especfico que a menudo es un desafo (o incluso imposible) en muchas aplicaciones. Esto condujo al desarrollo de muchas heursticas y enfoques simplificados que ayudaron a los analistas sin conocimiento del dominio a elegir el nmero apropiado de clsteres. </li>

Existe una gran cantidad de algoritmos de agrupamiento y, actualmente, no hay uno que domine a los dems. La eleccin del mejor depende de la base de datos en s, del dominio de la aplicacin y de los requisitos y expectativas del cliente.
Este cuaderno se centra en tres algoritmos de particin: K-Means, DBSCAN y Affinity Propagation. Todos estn implementados en una conocida biblioteca de Python: Scikit-Learn.

### 4.2 DBSCAN<a href='#Top' style="text-decoration: none;">^</a><br>  

En este cuaderno vamos a describir como DBSCAN (Density-Based Spatial Clustering of Applications with Noise) trabaja y mostrar su implementacin para mall customers segmentation.

<a id='DBSCAN description'></a>
### 4.2.1 Descripcin<a href='#Top' style="text-decoration: none;">^</a><br>  
DBSCAN stands for Density-Based Spatial Clustering of Applications with Noise, es un algoritmo de clustering

Como su nombre nos indica DBSCAN es un concepto de Regin Densas. Supongamos que un cluster natural esta compuesto punto locales densos. Esta definicion requiere de "Regin de densidad". Estos dos parmetros se requieren para el algortimo de DBSCAN

* Eps,   - distance
* MinPts  Minimum number of points within distance Eps  

obsionalmente la distancia mtrica la podemos especificar nosotros mismo, por defauld se usa la distancia Euclideana.


Una dense region es por tanto creado por un mnimo punto con distancia entre ellos mismos. Punto que estn en esta distancia pero no estn cerca del nmero mnimo de otros puntos se tratan como border points. Los restantes son ruidos o valores atpicos.

Ventajas:
* Encuentra el nmero de cluster en si, basado en Eps y Minpts parmetros.
* Es capaz de diferenciar entre grandes cluster o cluster rodeado de otros cluster
* Es capaz de encontrar puntos que nos estn dentro de algn cluster- decta outliers.

El mayor inconveniente de DBSCAN:
* Requiere un alto consumo promedio computacional, esto viene de la necesidad de ejecutar una gran cantidad de neighbourhood para cada punto.
* Es muy pobre para identificar cluster con varias densidades.


<a id='DBSCAN use case'></a>
### 4.2.2 Implemetacin DBSCAN - customers segmentation<a href='#Top' style="text-decoration: none;">^</a><br>  


```python
from sklearn.cluster import DBSCAN
```

Vamos a escoger la mejor combinacin de parmetro para el algoritmo. primero creamos una matriz para investigar las combinaciones.

```python
from itertools import product

eps_values = np.arange(8,12.75,0.25) # Valores eps a investigar
min_samples = np.arange(3,10) # Valores min_samples a investigar
DBSCAN_params = list(product(eps_values, min_samples))
```

```python
DBSCAN_params
```

Visualizamos el nmero de cluster generado.

```python
from sklearn import metrics
```

```python
X_numerics=mall_data.drop(["CustomerID", "Gender"], axis=1)
```

```python
no_of_clusters = []
sil_score = []

for p in DBSCAN_params:
    DBS_clustering = DBSCAN(eps=p[0], min_samples=p[1]).fit(X_numerics)
    no_of_clusters.append(len(np.unique(DBS_clustering.labels_)))
    sil_score.append(metrics.silhouette_score(X_numerics, DBS_clustering.labels_))
```

Mostramos los cluster generado por el algoritmo para cada combinacin de parmetros.

```python
tmp = pd.DataFrame.from_records(DBSCAN_params, columns =['Eps', 'Min_samples'])
tmp['No_of_clusters'] = no_of_clusters

pivot_1 = pd.pivot_table(tmp, values='No_of_clusters', index='Min_samples', columns='Eps')

fig, ax = plt.subplots(figsize=(12,6))
sns.heatmap(pivot_1, annot=True,annot_kws={"size": 16}, cmap="YlGnBu", ax=ax)
ax.set_title('Number of clusters')
plt.show()
```

En el heatplot muestra, el nmero de cluster que varia de 17 a 4.

```python
tmp = pd.DataFrame.from_records(DBSCAN_params, columns =['Eps', 'Min_samples'])
tmp['Sil_score'] = sil_score

pivot_1 = pd.pivot_table(tmp, values='Sil_score', index='Min_samples', columns='Eps')

fig, ax = plt.subplots(figsize=(18,6))
sns.heatmap(pivot_1, annot=True, annot_kws={"size": 10}, cmap="YlGnBu", ax=ax)
ax.set_title('Silhouette Score')
plt.show()
```

Global maximum es de  0.26 para un  eps=12.5 y min_samples=4.

```python
DBS_clustering = DBSCAN(eps=12.5, min_samples=4).fit(X_numerics)

DBSCAN_clustered = X_numerics.copy()
DBSCAN_clustered.loc[:,'Cluster'] = DBS_clustering.labels_ # append labels to points
```

Clusters sizes.

```python
DBSCAN_clust_sizes = DBSCAN_clustered.groupby('Cluster').size().to_frame()
DBSCAN_clust_sizes.columns = ["DBSCAN_size"]
DBSCAN_clust_sizes
```

DBSCAN crea 5 clusters ms un outliers cluster (-1). El tamao del clusters 0-4 varia significativamente - algunos solo tiene 4 o 8 observaciones. Hay 18 outliers.

```python
outliers = DBSCAN_clustered[DBSCAN_clustered['Cluster']==-1]

fig2, (axes) = plt.subplots(1,2,figsize=(12,5))


sns.scatterplot(x = 'Annual Income (k$)', y = 'Spending Score (1-100)',
                data=DBSCAN_clustered[DBSCAN_clustered['Cluster']!=-1],
                hue='Cluster', ax=axes[0], palette='Set1', legend='full', s=45)

sns.scatterplot(x = 'Age', y = 'Spending Score (1-100)',
                data=DBSCAN_clustered[DBSCAN_clustered['Cluster']!=-1],
                hue='Cluster', palette='Set1', ax=axes[1], legend='full', s=45)

axes[0].scatter(outliers['Annual Income (k$)'], outliers['Spending Score (1-100)'], s=5, label='outliers', c="k")
axes[1].scatter(outliers['Age'], outliers['Spending Score (1-100)'], s=5, label='outliers', c="k")
axes[0].legend()
axes[1].legend()

plt.setp(axes[0].get_legend().get_texts(), fontsize='10')
plt.setp(axes[1].get_legend().get_texts(), fontsize='10')

plt.show();
```

## 4.3 Affinity Propagation<a href='#Top' style="text-decoration: none;">^</a><br>  


**La propagacin de afinidad (AP)** es un algoritmo de agrupacin basado en el concepto de "paso de mensajes" entre puntos de datos. A diferencia de los algoritmos de agrupamiento como k-medias o k-medoides,la propagacin de afinidad no requiere que se determine o estime el nmero de clsteres antes de ejecutar el algoritmo. Similar a los k-medoides,  la propagacin de afinidad encuentra "ejemplares", miembros del conjunto de entrada que son representativos de los clsteres


<a id='AF description'></a>
### 4.3.1 Description<a href='#Top' style="text-decoration: none;">^</a><br>  
Affinity Propagation es un algoritmo de agrupacin propuesto por primera vez por Brendan Frey y Delbert Dueck en 2007 (<a href="https://science.sciencemag.org/content/315/5814/972"> "Agrupacin mediante el paso de mensajes entre Puntos de datos "</a>). Se basa en el concepto de enviar mensajes entre un par de puntos hasta que convergen. Estos mensajes son una forma de medir qu tan similares son estos dos puntos entre s y si pueden ser ejemplos entre s. El algoritmo encuentra un nmero ptimo de clsteres por s mismo. Esto tambin implica un costo de complejidad de tiempo muy alto del orden O (nT) donde n es el nmero de muestras y T es el nmero de iteraciones hasta la convergencia. Sin embargo, una gran ventaja de AP es la falta de sensibilidad a los criterios de inicializacin.
    
El usuario debe especificar dos parmetros:
* Preferencia que es un nmero negativo y controla cuntos ejemplos se utilizan
* Factor de amortiguacin que evita oscilaciones numricas al actualizar mensajes

Al igual que con cualquier otro algoritmo, se estn realizando esfuerzos para mejorarlo, por ejemplo:
* P.Ling y su equipo propusieron un algoritmo de <a href="https://www.sciencedirect.com/science/article/pii/S0167865516303403#bib0001"> "Propagacin de afinidad de preferencia ajustable (APAP)" </a> que puede producir mejores resultados de agrupamiento debido a la mejora de los clculos de preferencia de elementos
* H. Wenlong y su equipo propusieron <a href="https://www.sciencedirect.com/science/article/pii/S0020025516300561"> Transferencia de clsteres basados en la propagacin por afinidad </a> que supera al algoritmo actual en el caso de un conjunto de datos muy pequeo.

### 4.3.2 Algoritmo

Sea $x_1$ a $x_n$ un conjunto de puntos de datos, sin suposiciones sobre su estructura interna, y sea $s$ una funcin que cuantifique la similitud entre dos puntos cualesquiera, tal que $s(i, j) > s(i, k)$ iff $x_i$ es ms similar a $x_j$ que a $x_k$ . Para este ejemplo, se utiliz la distancia al cuadrado negativa de dos puntos de datos, es decir, para los puntos $x_i$ y $x_k$,  $s(i,k)=-\left\|x_{i}-x_{k}\right\|^{2}$.

La diagonal de $s$ (es decir, $s(i,i)$) es particularmente importante, ya que representa la preferencia de instancia, lo que significa la probabilidad de que una instancia en particular se convierta en un ejemplo. Cuando se establece en el mismo valor para todas las entradas, controla cuntas clases produce el algoritmo. Un valor cercano a la similitud mnima posible produce menos clases, mientras que un valor cercano o mayor que la mxima similitud posible produce muchas clases. Por lo general, se inicializa con la similitud mediana de todos los pares de entradas.

El algoritmo procede alternando entre dos pasos de paso de mensajes, que actualizan dos matrices:

* La matriz de "responsabilidad" $R$ tiene valores $r(i, k)$ que cuantifican qu tan adecuado es $x_k$ para servir como ejemplo para $x_i$, en relacin con otros ejemplos candidatos para $x_i$.
* La matriz de "disponibilidad" $A$ contiene valores a $(i, k)$ que representan cun "apropiado" sera para $x_i$ elegir $x_k$ como su ejemplo, teniendo en cuenta la preferencia de otros puntos por $x_k$ como ejemplo.

Ambas matrices se inicializan a todos los ceros y se pueden ver como tablas de probabilidad logartmica. A continuacin, el algoritmo realiza las siguientes actualizaciones de forma iterativa:

Primero, se envan actualizaciones de responsabilidad: $${\displaystyle r(i,k)\leftarrow s(i,k)-\max _{k'\neq k}\left\{a(i,k')+s(i,k')\right\}}$$

A continuacin, la disponibilidad se actualiza por
$${\displaystyle a(i,k)\leftarrow \min \left(0,r(k,k)+\sum _{i'\not \in \{i,k\}}\max(0,r(i',k))\right)} para \ {\displaystyle i\neq k} $$

$${\displaystyle a(k,k)\leftarrow \sum _{i'\neq k}\max(0,r(i',k))}.$$

Las iteraciones se realizan hasta que los lmites del clster permanecen sin cambios durante varias iteraciones o se alcanza algn nmero predeterminado (de iteraciones). Los ejemplares se extraen de las matrices finales como aquellos cuya "responsabilidad + disponibilidad" para s mismos es positiva (es decir, $(r(i,i)+a(i,i))>0$).

```python
import numpy as np
preferences = range(-20000,-5000,99) # arbitraty chosen range    -    #does not work on -14700
no_of_clusters = []
af_sil_score = [] # silouette scores

for p in preferences:
    AF = AffinityPropagation(preference=p, max_iter=200).fit(X_numerics)
    no_of_clusters.append((len(np.unique(AF.labels_))))
    af_sil_score.append(metrics.silhouette_score(X_numerics, AF.labels_))

af_results = pd.DataFrame([preferences, no_of_clusters, af_sil_score], index=['Preference','Num. of Clusters','Silhouette Score']).T
af_results = af_results.sort_values(by='Silhouette Score', ascending=False).reset_index(drop=True)
af_results
```
