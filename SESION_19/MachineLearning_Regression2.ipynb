# **Supervised Learning modelo de Regresin mediante  Sklearn**

El aprendizaje automtico es una rama de la inteligencia artificial que permite a los programas informticos aprender y mejorar automticamente a partir de la experiencia. Los algoritmos de aprendizaje automtico aprenden de los conjuntos de datos y, luego, basndose en los patrones identificados en los conjuntos de datos, hacen predicciones sobre datos invisibles.

Los algoritmos de aprendizaje automtico se pueden clasificar principalmente en dos tipos: algoritmos de aprendizaje supervisado y algoritmos de aprendizaje no supervisado.

Los algoritmos de aprendizaje automtico supervisados son aquellos algoritmos en los que el conjunto de datos de entrada y la salida correspondiente o la prediccin verdadera estn disponibles, y los algoritmos intentan encontrar la relacin entre las entradas y las salidas.

Los algoritmos de aprendizaje supervisado se dividen en dos tipos: algoritmos de regresin y algoritmos de clasificacin.

Los algoritmos de regresin predicen un valor continuo, por ejemplo, el precio de una casa, la presin arterial de una persona y la puntuacin de un estudiante en un examen en particular. Los algoritmos de clasificacin, por otro lado, predicen un valor discreto, como si un tumor es maligno o no, si un estudiante va a aprobar o reprobar un examen, etc.

## **2. Regresin lineal**

La regresin lineal es un modelo lineal que supone una relacin lineal entre las entradas y las salidas y minimiza el costo del error entre la salida prevista y la real utilizando funciones como el error absoluto medio entre diferentes puntos de datos.

En este apartado, veremos los modelos de aprendizaje automtico supervisado ms comunes. Los algoritmos de aprendizaje automtico supervisado funcionan de manera diferente a los modelos de series temporales.

En los modelos de aprendizaje automtico supervisado, intenta identificar relaciones entre diferentes variables:

* Variable objetivo: la variable que intenta pronosticar
* Variables explicativas: Variables que te ayudan a predecir la variable objetivo

Para la previsin, es importante comprender qu tipos de variables explicativas puede o no utilizar. Como ejemplo, digamos que las ventas de chocolate caliente dependen fuertemente de la temperatura. Cuando el clima es fro, las ventas son altas. Cuando el clima es clido, las ventas son bajas.





## **La idea detrs de la regresin lineal**
La idea detrs de la regresin lineal es definir una relacin lineal entre una variable objetivo y numerosas variables explicativas para predecir la variable objetivo. La regresin lineal se usa ampliamente, no solo para pronosticar. Como cualquier modelo supervisado, siempre y cuando coloque variables explicativas del futuro como entrada al modelo, esto funciona perfectamente.

## **Definicin del modelo**
La regresin lineal se define de la siguiente manera:

$$y=\beta_0+\beta_1 x_1+\beta_2 x_2+\cdots+\beta_p x_p $$

En esta frmula
* Existen $p$ variables explicativas, denominadas $x$.
* Hay una variable objetivo llamada $y$.
* El valor de $y$ se calcula como una constante ($\beta_0$) ms los valores de las variables $x$ multiplicados por sus coeficientes $\beta_1$ a $\beta_p$.

La figura de abajo se muestra cmo interpretar $\beta_0$ y $\beta_1$ visualmente. Muestra que para un aumento de 1 en la variable $x$, el aumento en la variable y representa $\beta_1$. $\beta_0$ es el valor de $y$ cuando $x$ es 0.

Para poder usar la regresin lineal, debe estimar los coeficientes (betas) en un conjunto de datos de entrenamiento. Luego, los coeficientes se pueden estimar utilizando la siguiente frmula, en notacin matricial:

$$\hat \beta=(X^T X)^{-1} X^T y $$

Esta frmula se conoce como MCO: el mtodo de los mnimos cuadrados ordinarios. Este modelo es muy rpido de ajustar, ya que solo requiere clculos matriciales para calcular las betas. Aunque es fcil de instalar, es menos adecuado para procesos ms complejos. Despus de todo, es un modelo lineal y, por lo tanto, solo puede ajustarse a procesos lineales.

Un modelo lineal puede ajustarse a cualquier tipo de relacin que vaya en una direccin. Un ejemplo de esto es si sube la temperatura, bajan las ventas de chocolate caliente. Sin embargo, los modelos lineales no pueden adaptarse a nada no lineal. Un ejemplo de un proceso no lineal es este: si la temperatura est bajo cero, las ventas de chocolate caliente son bajas; si la temperatura est entre cero y 10, las ventas de chocolate caliente son altas; si la temperatura es alta, las ventas de chocolate caliente son bajas.

Como puede ver, el segundo ejemplo no es lineal porque no se podra trazar una lnea recta desde las ventas bajas hasta las altas de chocolate caliente. Ms bien, podra hacer mejor una declaracin if/else para capturar esta lgica.

Debe tener en cuenta que los modelos lineales no son muy buenos para capturar tendencias no lineales. Sin embargo, los modelos no lineales, cuando se ajustan correctamente, a menudo pueden aproximarse bastante bien a las tendencias lineales. Esta es la razn por la que muchas de las tcnicas de aprendizaje automtico ms avanzadas utilizan muchos enfoques no lineales.

### **2.7 Tipos de Metricas**

Una vez que haya entrenado un modelo y haya hecho predicciones en el conjunto de prueba, el siguiente paso es saber qu tan bien se ha desempeado su modelo para hacer predicciones en el conjunto de prueba desconocido. Hay varias mtricas para comprobarlo. Sin embargo, el error absoluto medio, el error cuadrtico medio y el error cuadrtico medio medio son tres de las mtricas ms comunes.

**R2**

La mtrica R2 (R cuadrado) es una mtrica que est muy cerca de la mtrica 1 - MAPE. Es una mtrica de rendimiento en lugar de una mtrica de error, lo que la hace ideal para la comunicacin.

El R2 es un valor que tiende a estar entre 0 y 1, siendo 0 malo y 1 perfecto. Por lo tanto, se puede usar fcilmente como un porcentaje multiplicndolo por 100. El nico caso en el que el R2 puede ser negativo es si su pronstico es ms del 100 % incorrecto.

$$R^2=\frac{\sum(y_i -\hat y_i)^2 }{\sum (y_i -\bar y_i )^2} $$

La frmula hace un clculo interesante. Calcula una relacin entre la suma de los errores al cuadrado y la suma de las desviaciones entre el pronstico y el promedio. Esto se reduce a un porcentaje de aumento de su modelo sobre el uso del promedio como modelo. Si su modelo es una prediccin tan mala como usar el promedio, entonces el R2 ser cero. Como el promedio se usa a menudo como modelo de referencia, esta es una mtrica de rendimiento muy prctica.

**Error absoluto medio(Mean Absolute Error MAE)**

El error absoluto medio (MAE) se calcula tomando el promedio del error absoluto obtenido al restar los valores reales de los valores predichos. La ecuacin para calcular MAE es:

$$MAE=\frac{\Sigma_{i=1}^{n} |y_i-\hat{y}|}{n} $$

**Error medio cuadrado (Mean Squared Error MSE)**

El error cuadrtico medio (MSE) es similar al MAE. Sin embargo, el error para cada registro se eleva al cuadrado en el caso de MSE para castigar los registros de datos con una gran diferencia entre los valores predichos y reales. La ecuacin para calcular el error cuadrtico medio es la siguiente:

$$MSE=\frac{1}{n} \Sigma_{i=1}^{n} (y_i-\hat{y})^2 $$

**Root Mean Squared Error**

El error cuadrtico medio es simplemente la raz inferior del error cuadrtico medio y se puede calcular de la siguiente manera:

$$RMSE=\sqrt{\frac{1}{n} \Sigma_{i=1}^{n} (y_i-\hat{y})^2  } $$

Los mtodos utilizados para encontrar el valor de estas mtricas estn disponibles en la clase `sklearn.metrics`. Los valores predichos y reales deben pasarse a estos mtodos, como se muestra en el resultado.

```python
from sklearn import metrics

print ("Mean Absolute Error:" , metrics.mean_absolute_error(y_test, y_pred_lr))
print ("Mean Squared Error:" , metrics.mean_squared_error(y_test, y_pred_lr))
RMSE_lr = np.sqrt(metrics.mean_squared_error(y_test, y_pred_lr))
print ("Root Mean Squared Error:" , RMSE_lr)
# The coefficient of determination: 1 is perfect prediction
R2_lr = r2_score(y_test, y_pred_lr)
print('Coeficiente de determinacin: %.2f'% R2_lr)
```

##  **3. Modelo KNN para Regresin**

El KNN significa K-nearest neighbors. KNN es un algoritmo de aprendizaje automatico, que se basa en encontrar la distancia euclidiana entre diferentes puntos de datos.

```python
#@title Algoritmo KNN
from IPython.display import Image
Image("/content/drive/MyDrive/Images/knn.jpg")
```

**Los pasos que se siguen en k-vecino ms cercano son los siguientes:**
1. Recibimos el conjunto de datos sin procesar y sin clasificar que hay que trabajar.
2. Elegimos una matriz de distancias de la euclidiana, Manhattan o Minkowski.
3. Luego calcule la distancia entre los nuevos datos puntos y los puntos de datos de entrenamiento clasificados conocidos.
4. El nmero de vecinos a considerar es definida por el valor de k.
5. Se sigue comparando con la lista de clases. que tienen la distancia ms corta y cuentan las nmero de veces que aparece cada clase.
6. La clase con la mayor cantidad de votos gana. Esto significa que la clase que tiene la frecuencia ms alta y ha aparecido el mayor nmero de veces es asignado al punto de datos desconocido.

**Mtodos de clculo de la distancia entre puntos**
El primer paso es calcular la distancia entre el nuevo punto y cada punto de entrenamiento. Existen varios mtodos para calcular esta distancia, de los cuales los mtodos ms comnmente conocidos son: euclidiano, Manhattan (para continuo) y distancia de Hamming (para categrico).

1. Distancia euclidiana: probablemente la forma ms comn y sencilla de calcular entre dos puntos. Es raz cuadrada de la suma de los cuadrados de las distancias:
$$Euclidean=\sqrt{(y_2-y_1)^2+(x_2-x_1)^2} $$

2. **Distancia de Manhattan:** La distancia entre dos puntos medida a lo largo de ejes en ngulo recto. A veces tambin se le conoce como distancia de cuadra de la ciudad. En un plano con $p_1$ en $(x_1, y_1)$ y $p_2$ en $(x_2, y_2)$, es

$$Manhattan=|x_1-x_2|+|y_1-y_2| $$

3. **Distancia de Minkowski:** Esta es una mtrica en un espacio vectorial normado. La distancia de Minkowski se utiliza para la similitud de distancia de los vectores. Dados dos o ms vectores, encuentre la similitud de distancia de estos vectores. Principalmente, la distancia de Minkowski se aplica en ML para averiguar la similitud de la distancia. Es una mtrica de distancia generalizada y se puede representar mediante la siguiente frmula:

$$Minkowski={(\sum_{i=1}^{n}|x_i-y_i|^p )^{1/p} } $$

donde usando diferentes valores de p podemos obtener diferentes valores de distancias. Con el valor de $p = 1$ tenemos obtener la distancia de Manhattan, con $p = 2$ obtenemos euclidiana distancia, y con $p= $ obtenemos la distancia de Chebychev (https://www.youtube.com/watch?v=XP5L5mrUaDw).

4. **Cosine Similarity:** Es una medida de semejanza entre dos vectores distintos de cero de un producto interno espacio que mide el coseno del ngulo entre ellos. El coseno de 0 es 1, y es menor que 1 para cualquier ngulo en el intervalo $(0,]$ radianes.

El error absoluto medio (MAE) se calcula tomando el promedio del error absoluto obtenido al restar los valores reales de los valores predichos. La ecuacin para calcular MAE es:

**Error medio cuadrado (Mean Squared Error MSE)**

El error cuadrtico medio (MSE) es similar al MAE. Sin embargo, el error para cada registro se eleva al cuadrado en el caso de MSE para castigar los registros de datos con una gran diferencia entre los valores predichos y reales. La ecuacin para calcular el error cuadrtico medio es la siguiente:

$$MSE=\frac{1}{2} \Sigma_{i=1}^{n} (y_i-\hat{y})^2 $$

**Root Mean Squared Error**

El error cuadrtico medio es simplemente la raz inferior del error cuadrtico medio y se puede calcular de la siguiente manera:

$$RMSE=\sqrt{\frac{1}{n} \Sigma_{i=1}^{n} (y_i-\hat{y})^2  } $$

Sigamos adelante y usemos el mtodo del codo para elegir un buen valor $K$. Bsicamente, comprobaremos la tasa de error de $k = 1$ para decir $k = 20$. Para cada valor de $k$ llamaremos al Regresor de  KNN y luego elegiremos el valor de $k$ que tenga la menor tasa de error.

```python
from sklearn import neighbors
from sklearn.metrics import mean_squared_error
from math import sqrt
import matplotlib.pyplot as plt
%matplotlib inline

rmse_val = [] #para almacenar valores rmse para diferentes k
for K in range(20):
    K = K+1
    model = neighbors.KNeighborsRegressor(n_neighbors = K)

    model.fit(X_train, y_train)  #fit the model
    pred=model.predict(X_test) #hacer predicciones en el equipo de prueba
    error = sqrt(mean_squared_error(y_test,pred)) #calculate rmse
    rmse_val.append(error) #store rmse values
    print('Valor de RMSE para k= ' , K , 'es:', error)
```

El error absoluto medio (MAE) se calcula tomando el promedio del error absoluto obtenido al restar los valores reales de los valores predichos. La ecuacin para calcular MAE es:

$$MAE=\frac{\Sigma_{i=1}^{n} |y_i-\hat{y}|}{n} $$

**Error medio cuadrado (Mean Squared Error MSE)**

### **5.1. Introduccin al algoritmo del rbol de decisiones**

Un algoritmo de rbol de decisiones es uno de los algoritmos de aprendizaje automtico ms populares. Utiliza una estructura similar a un rbol y sus posibles combinaciones para resolver un problema en particular. Pertenece a la clase de algoritmos de aprendizaje supervisado donde se puede utilizar tanto para fines de clasificacin como de regresin.


Un rbol de decisin es una estructura que incluye un nodo raz, ramas y nodos hoja. Cada nodo interno denota una prueba en un atributo, cada rama denota el resultado de una prueba y cada nodo hoja contiene una etiqueta de clase. El nodo superior del rbol es el nodo raz.


Hacemos algunas suposiciones al implementar el algoritmo Decision-Tree. Estos se enumeran a continuacin: -

1. Al principio, todo el conjunto de entrenamiento se considera como raz.
2. Los valores de las caractersticas deben ser categricos. Si los valores son continuos, se discretizan antes de construir el modelo.
3. Los registros se distribuyen recursivamente sobre la base de los valores de los atributos.
4. El orden para colocar atributos como raz o nodo interno del rbol se realiza mediante algn enfoque estadstico.

### **5.2. rboles de clasificacin y regresin (CART)**

Hoy en da, el algoritmo del rbol de decisiones se conoce por su nombre moderno **CART**, que significa **rboles de clasificacin y regresin**. Los rboles de clasificacin y regresin o **CART** es un trmino introducido por Leo Breiman para referirse a los algoritmos de rboles de decisin que se pueden usar para problemas de modelado de clasificacin y regresin.


El algoritmo CART proporciona una base para otros algoritmos importantes, como rboles de decisin empaquetados, bosques aleatorios y rboles de decisin potenciados. En este ncleo, resolver un problema de clasificacin. Por lo tanto, me referir al algoritmo tambin como problema de Clasificacin del rbol de Decisin.
